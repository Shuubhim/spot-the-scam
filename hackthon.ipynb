```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings('ignore')

# ===============================
# 1. LOAD AND INSPECT DATA
# ===============================

# Load your datasets
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print("=== DATASET OVERVIEW ===")
print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
print(f"Target distribution: {train_df['fraudulent'].value_counts(normalize=True)}")

# ===============================
# 2. HANDLE MISSING VALUES
# ===============================

def handle_missing_values(df):
    """Smart missing value handling"""

    # Drop columns with >70% missing values
    high_missing_cols = ['salary_range', 'department']  # 84% and 65% missing
    df = df.drop(columns=high_missing_cols, errors='ignore')

    # Fill text columns with 'missing'
    text_cols = ['description', 'requirements', 'benefits', 'company_profile', 'title']
    for col in text_cols:
        if col in df.columns:
            df[col] = df[col].fillna('missing')

    # Fill categorical columns with 'Not Provided'
    categorical_cols = ['employment_type', 'required_experience', 'required_education',
                       'industry', 'function', 'location']
    for col in categorical_cols:
        if col in df.columns:
            df[col] = df[col].fillna('Not Provided')

    return df

# Apply missing value handling
train_df = handle_missing_values(train_df)
test_df = handle_missing_values(test_df)

# ===============================
# 3. FEATURE ENGINEERING
# ===============================

def create_text_features(df):
    """Create text-based features"""

    # Text length features
    df['desc_length'] = df['description'].str.len()
    df['title_length'] = df['title'].str.len()
    df['requirements_length'] = df['requirements'].str.len()

    # Text quality indicators
    df['has_description'] = (df['description'] != 'missing').astype(int)
    df['has_requirements'] = (df['requirements'] != 'missing').astype(int)
    df['has_benefits'] = (df['benefits'] != 'missing').astype(int)

    return df

def clean_text(text):
    """Clean text data"""
    if pd.isna(text) or text == 'missing':
        return ''

    # Remove HTML tags
    text = re.sub(r'<[^>]+>', '', text)
    # Remove special characters and extra spaces
    text = re.sub(r'[^\w\s]', ' ', text)
    # Convert to lowercase
    text = text.lower().strip()
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)

    return text

def engineer_features(df):
    """Create all engineered features"""

    # Create text features
    df = create_text_features(df)

    # Clean text columns
    text_columns = ['description', 'title', 'requirements', 'benefits', 'company_profile']
    for col in text_columns:
        if col in df.columns:
            df[f'{col}_clean'] = df[col].apply(clean_text)

    # Location features
    if 'location' in df.columns:
        df['country'] = df['location'].str.split(',').str[0].str.strip()
        df['is_us'] = (df['country'] == 'US').astype(int)

    # Binary feature combinations
    df['logo_and_questions'] = df['has_company_logo'] * df['has_questions']
    df['remote_no_logo'] = df['telecommuting'] * (1 - df['has_company_logo'])

    return df

# Apply feature engineering
train_df = engineer_features(train_df)
test_df = engineer_features(test_df)

# ===============================
# 4. ENCODE CATEGORICAL VARIABLES
# ===============================

def encode_categorical_features(train_df, test_df):
    """Encode categorical variables"""

    # Get categorical columns (excluding target and text columns)
    categorical_cols = ['employment_type', 'required_experience', 'required_education',
                       'industry', 'function', 'country']

    # Filter existing columns
    categorical_cols = [col for col in categorical_cols if col in train_df.columns]

    # One-hot encode categorical variables
    train_encoded = pd.get_dummies(train_df, columns=categorical_cols, dummy_na=True)
    test_encoded = pd.get_dummies(test_df, columns=categorical_cols, dummy_na=True)

    # Align columns between train and test
    train_cols = set(train_encoded.columns)
    test_cols = set(test_encoded.columns)

    # Add missing columns to test set
    for col in train_cols - test_cols:
        if col != 'fraudulent':  # Don't add target to test
            test_encoded[col] = 0

    # Add missing columns to train set (shouldn't happen but safety check)
    for col in test_cols - train_cols:
        train_encoded[col] = 0

    # Reorder columns to match
    common_cols = [col for col in train_encoded.columns if col in test_encoded.columns]
    train_encoded = train_encoded[common_cols + ['fraudulent'] if 'fraudulent' in train_encoded.columns else common_cols]
    test_encoded = test_encoded[common_cols]

    return train_encoded, test_encoded

# Apply encoding
train_processed, test_processed = encode_categorical_features(train_df, test_df)

# ===============================
# 5. PREPARE TEXT FEATURES
# ===============================

def prepare_text_features(train_df, test_df, max_features=5000):
    """Create TF-IDF features from text"""

    # Combine description and title for richer text features
    train_df['combined_text'] = (train_df['description_clean'].fillna('') + ' ' +
                                train_df['title_clean'].fillna('') + ' ' +
                                train_df['requirements_clean'].fillna(''))

    test_df['combined_text'] = (test_df['description_clean'].fillna('') + ' ' +
                               test_df['title_clean'].fillna('') + ' ' +
                               test_df['requirements_clean'].fillna(''))

    # Create TF-IDF features
    tfidf = TfidfVectorizer(
        max_features=max_features,
        stop_words='english',
        ngram_range=(1, 2),  # Include bigrams
        min_df=2,
        max_df=0.95
    )

    # Fit on training data and transform both sets
    train_tfidf = tfidf.fit_transform(train_df['combined_text'])
    test_tfidf = tfidf.transform(test_df['combined_text'])

    # Convert to DataFrame
    tfidf_cols = [f'tfidf_{i}' for i in range(train_tfidf.shape[1])]
    train_tfidf_df = pd.DataFrame(train_tfidf.toarray(), columns=tfidf_cols, index=train_df.index)
    test_tfidf_df = pd.DataFrame(test_tfidf.toarray(), columns=tfidf_cols, index=test_df.index)

    return train_tfidf_df, test_tfidf_df

# Create TF-IDF features
train_tfidf, test_tfidf = prepare_text_features(train_df, test_df)

# ===============================
# 6. COMBINE ALL FEATURES
# ===============================

# Select numeric and binary features
numeric_features = ['desc_length', 'title_length', 'requirements_length',
                   'telecommuting', 'has_company_logo', 'has_questions',
                   'has_description', 'has_requirements', 'has_benefits',
                   'logo_and_questions', 'remote_no_logo', 'is_us']

# Filter existing features
numeric_features = [col for col in numeric_features if col in train_processed.columns]

# Combine all features
X_train = pd.concat([
    train_processed[numeric_features],
    train_tfidf
], axis=1)

X_test = pd.concat([
    test_processed[numeric_features],
    test_tfidf
], axis=1)

y_train = train_processed['fraudulent']

# ===============================
# 7. FINAL SUMMARY
# ===============================

print("\n=== PREPROCESSING COMPLETE ===")
print(f"Final training set shape: {X_train.shape}")
print(f"Final test set shape: {X_test.shape}")
print(f"Target distribution: {y_train.value_counts(normalize=True)}")
print(f"Features created: {X_train.columns.tolist()[:10]}... (showing first 10)")

# Save processed data
X_train.to_csv('X_train_processed.csv', index=False)
X_test.to_csv('X_test_processed.csv', index=False)
y_train.to_csv('y_train.csv', index=False)

print("\n✅ Preprocessed data saved to CSV files!")
print("Next step: Model training and evaluation")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Load preprocessed data
X = pd.read_csv('X_train_processed.csv')
y = pd.read_csv('y_train.csv').values.ravel()

# Split into train/validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Calculate class weights
pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  # ~19.6

# ======================
# 1. XGBoost Model
# ======================
xgb = XGBClassifier(
    scale_pos_weight=pos_weight,
    eval_metric='logloss',
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    early_stopping_rounds=10,
    random_state=42
)

# Train with early stopping
xgb.fit(X_train, y_train,
        eval_set=[(X_val, y_val)],
        verbose=True)

# Feature Importance
plt.figure(figsize=(10, 6))
sorted_idx = xgb.feature_importances_.argsort()
plt.barh(X.columns[sorted_idx][-15:], xgb.feature_importances_[sorted_idx][-15:])
plt.xlabel("XGBoost Feature Importance")
plt.show()

# ======================
# 2. Random Forest Model
# ======================
rf = RandomForestClassifier(
    class_weight='balanced',
    n_estimators=200,
    max_depth=8,
    random_state=42
)

rf.fit(X_train, y_train)

# Feature Importance
plt.figure(figsize=(10, 6))
sorted_idx = rf.feature_importances_.argsort()
plt.barh(X.columns[sorted_idx][-15:], rf.feature_importances_[sorted_idx][-15:])
plt.xlabel("Random Forest Feature Importance")
plt.show()

# ======================
# 3. Logistic Regression
# ======================
logreg = LogisticRegression(
    class_weight='balanced',
    penalty='l2',
    solver='liblinear',
    max_iter=1000,
    random_state=42
)

logreg.fit(X_train, y_train)

# ======================
# Model Evaluation
# ======================
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]

    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print(f"ROC AUC: {roc_auc_score(y_test, y_proba):.3f}")
    print(f"F1 Score: {f1_score(y_test, y_pred):.3f}")

    # Confusion Matrix (Percentages)
    cm = confusion_matrix(y_test, y_pred)
    cm_percent = cm / cm.sum(axis=1)[:, np.newaxis]

    plt.figure(figsize=(6,4))
    sns.heatmap(cm_percent, annot=True, fmt='.1%', cmap='Blues',
                xticklabels=['Real', 'Fraud'], yticklabels=['Real', 'Fraud'])
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

print("\n=== XGBoost Evaluation ===")
evaluate_model(xgb, X_val, y_val)

print("\n=== Random Forest Evaluation ===")
evaluate_model(rf, X_val, y_val)

print("\n=== Logistic Regression Evaluation ===")
evaluate_model(logreg, X_val, y_val)

# ======================
# Model Saving
# ======================
joblib.dump(xgb, 'xgboost_model.pkl')
joblib.dump(rf, 'random_forest_model.pkl')
joblib.dump(logreg, 'logistic_regression.pkl')
```
